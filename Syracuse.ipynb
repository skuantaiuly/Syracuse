# The Syracuse Sequence

The Syracuse (Collatz or Hailstone) conjecture concerns a sequence of integers defined as follows:

    ‚ä≥ Starts with any positive integer N
        ‚ä≥ And each term is formed from the previous one as follows:
        
            ‚ä≥ if the previous term is odd:
                the next term is 3 times the previous term and plus 1 
$$
                3*N + 1
$$
            ‚ä≥ if the previous term is even:
                the next term is half the previous term
$$ 
                N / 2
$$
    
    The sequence will always reach 1. üèÅ


```python
import seaborn as sb
import matplotlib.pyplot as plt

def main(n):
    y = []

    number = n
    y.append(int(number))

    while number>1:
        # Syracuse sequence calculation
        if number%2==1: #if number is odd
            number=number*3+1
        elif number%2==0: #if number is even
            number=number/2

        y.append(int(number))

    print(y)

    fig, ax = plt.subplots(figsize=(15,5))
    sb.lineplot(y, marker='o', ax=ax).set(title='Line Graph visualize Syracuse sequence')
    plt.show()

main(27)
```
